```python
import os #talks to the OS to create, modify or delete folders
import glob # searches for files 
import shutil
```

# Task 1

```python
# split the string using sep'-'
# pick the first thing 
# and use the dictionary to convert 

conversion= {
'jan': '01', 'feb': '02', 'mar': '03',
'apr': '04', 'may': '05', 'jun': '06',
'jul': '07', 'aug': '08', 'sep': '09',
'oct': '10', 'nov': '11', 'dec': '12'
}
#split the 
file = 'oct-08_singapore.txt'

day, city = file.split('_') #unpacking gets rid of the list 
                            # oct-08, singapore 
print(day, city)

#split via the - 
# to get the month and the date
month, date = day.split('-')
print(month, date)



print(f'{conversion[month]}-{date}-{city}')
```

# Task 2

```python
def rename_my_file(old_file_name):
    day, city = old_file_name.split('_')
    month, date = day.split('-')
    return f'{conversion[month]}-{date}-{city}'
```

```python
rename_my_file('apr-01_shanghai.txt')
```

# Task 4

```python
# find all txt files in the inventory, havent renamed them yet
txt_files = glob.glob('os-collaboration-exercise-data//**/*.txt', recursive=True)
print(txt_files)
```

```python
for terxt in txt_files:                                 #for the files found in text files
    filetextsplit = terxt.split(os.path.sep)[-1]        #split the text files you found and pull out the last bit#
    renamedtextfiles = rename_my_file(filetextsplit) 
    joiningpath = os.path.join('os-collaboration-exercise-data\\os-collaboration-exercise-data', renamedtextfiles)   #the part where is says the date and city 
    os.rename(terxt, joiningpath)                  #
```

```python
cairo_files = glob.glob('os-collaboration-exercise-data/**/*[Cc]airo*', recursive=True)
print(cairo_files)
```

```python
for cairo_filez in cairo_files: 
    path_that_it_takes = os.path.join('Cities', 'Cairo')
    shutil.copy(cairo_filez, path_that_it_takes)
    print(f'all done')
```

# Task 7

```python
#loop thru each city 
for cityy in ['Cairo', 'Dhaka', 'Jakarta', 'Karachi', 'Manila',
 'Melbourne', 'Osaka', 'Shanghai', 'Singapore', 'Tokyo']: 
    path = os.path.join('Cities', cityy)
    #find city files, ensure path is a directory 
    find_city_files = glob.glob(f'os-collaboration-exercise-data/**/*{cityy}*', recursive=True)
   # previously, path = cities, cityy 
   # now directing files in the city files to the path
    for file in find_city_files: 
      filename = file.split('\\')[-1]
      target_path = os.path.join(path, filename)
      shutil.move(file, target_path)        

print('ok all done')
```
