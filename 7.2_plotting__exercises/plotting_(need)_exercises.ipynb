```python
import numpy as np
```

```python
from matplotlib import pyplot as plt
```

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5]
y = [0, 4, 8, 12, 16, 20]
y_err = [0.0, 0.5, 1.0, 1.5, 2.0, 2.5]
x_err = .25

plt.errorbar(x, y, 
             xerr= 0.25 , yerr= [0.0, 0.5, 1.0, 1.5, 2.0, 2.5],
             color= 'brown', linestyle= 'dashed',
             marker= 's', markerfacecolor='yellow',
             ecolor='black',capsize=5,
             label= 'Y values with errors')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc = 'lower right')
plt.show()
```

```python
x_1 = np.linspace(0, 50, 500)
y_2 = np.e**((-x_1)/10) * np.sin(x_1)


plt.plot(x_1, y_2, color = "tab:blue")
plt.ylim([-1, 1])
plt.xlabel('x', color= 'darkorange', fontsize = 15)
plt.ylabel('y', color= 'darkorange', fontsize = 15, rotation=0)
plt.grid(alpha=0.25)
plt.title('$y = e^{-x/10}sin(x)$', fontsize = 20)
```
